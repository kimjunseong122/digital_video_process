```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../images/char.bmp')
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
plt.imshow(dilateImg, cmap='gray')
```

```python
edgeImg = dilateImg - gImg
plt.imshow(edgeImg, cmap='gray')
```

```python
img1 = cv2.imread('../images/milkdrop.bmp')
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 5)
plt.imshow(dilateImg, cmap='gray')
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype= np.ubyte)

threshhold = 150

for h in range(gImg.shape[0]):
    for w in range(gImg.shape[1]):
        if gImg[h, w] > threshhold:
            gBin[h, w] = 255

plt.imshow(gBin, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(gBin, kernel, iterations = 3)
plt.imshow(dilateImg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
erodeImg = cv2.erode(gBin, kernel, iterations = 2)
plt.imshow(erodeImg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(erodeImg, kernel, iterations = 5)
plt.imshow(dilateImg, cmap='gray')
```
