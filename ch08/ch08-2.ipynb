```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
scanImg = cv2.imread('../images/Erosion.png')
RGB_img = cv2.cvtColor(scanImg, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

###   erode() (dilate)
### dilate() (erode)
###     

```python
kernel = np.ones((3, 3), np.uint8)
erosionImg = cv2.erode(gImg, kernel, iterations = 1)
plt.imshow(erosionImg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
plt.imshow(dilateImg, cmap='gray')
```

```python
kernel = np.ones((4, 1), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
plt.imshow(dilateImg, cmap='gray')
```

```python
kernel = np.ones((4, 1), np.uint8)
erosionImg = cv2.erode(dilateImg, kernel, iterations = 1)
plt.imshow(erosionImg, cmap='gray')
```
