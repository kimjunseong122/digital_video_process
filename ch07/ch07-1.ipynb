```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../images/shrinking.bmp')
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
# outImg = cv2.resize(gImg, None, fx = 0.5, fy = 0.5, interpolation=cv2.INTER_LINEAR)
# outImg = cv2.resize(gImg, None, fx = 0.3, fy = 0.3, interpolation=cv2.INTER_NEAREST)
outImg = cv2.resize(gImg, None, fx = 0.5, fy = 0.5, interpolation=cv2.INTER_CUBIC)

plt.imshow(outImg, cmap='gray')
```

```python
blurImg = cv2.GaussianBlur(gImg, (5, 5), 1)
outImg = cv2.resize(blurImg, None, fx = 0.5, fy = 0.5, interpolation=cv2.INTER_LINEAR)
# outImg = cv2.resize(blurImg, None, fx = 0.3, fy = 0.3, interpolation=cv2.INTER_NEAREST)
# outImg = cv2.resize(blurImg, None, fx = 0.3, fy = 0.3, interpolation=cv2.INTER_CUBIC)

plt.imshow(outImg, cmap='gray')
```
