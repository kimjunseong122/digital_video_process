```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel5 = np.ones((5,5), np.float32)/25
```

```python
# Low Frequency Filter
kernel3[0][0] = 1/9
kernel3[0][1] = 1/9
kernel3[0][2] = 1/9
kernel3[1][0] = 1/9
kernel3[1][1] = 1/9
kernel3[1][2] = 1/9
kernel3[2][0] = 1/9
kernel3[2][1] = 1/9
kernel3[2][2] = 1/9
```

```python
# Original Frequency Filter
kernel3[0][0] = 0
kernel3[0][1] = 0
kernel3[0][2] = 0
kernel3[1][0] = 0
kernel3[1][1] = 1
kernel3[1][2] = 0
kernel3[2][0] = 0
kernel3[2][1] = 0
kernel3[2][2] = 0
```

```python
# High Frequency Filter
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] = 8/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```

```python
outImg3 = cv2.filter2D(gImg, -1, kernel3)
# outImg3 = outImg3 * 10
plt.imshow(outImg3, cmap='gray')
```

```python
outImg5 = cv2.filter2D(gImg, -1, kernel5)
plt.imshow(outImg5, cmap='gray')
```

```python
outImg33 = cv2.filter2D(outImg3, -1, kernel3)
plt.imshow(outImg33, cmap='gray')
```

```python
outImg333 = cv2.filter2D(outImg3, -1, kernel3)
plt.imshow(outImg333, cmap='gray')
```
