```python
img = cv2.imread('../images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')

# cv2.imshow('Lena', RGB_img)
# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
gPlus =  np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)
```

```python
val = 100
gImg = gImg.astype(int)
for h in range(gImg.shape[0]):
    for w in range(gImg.shape[1]):
        plusVal = np.int32(gImg[h, w] + val)
        if(plusVal > 255):
            plusVal = 255
        gPlus[h, w] = plusVal

plt.imshow(gPlus, cmap='gray')
```
