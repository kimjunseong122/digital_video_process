```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/my_dog_jelly.jpeg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg1.shape)
plt.imshow(gImg1, cmap='gray')
```

```python
gOutimg = np.zeros((gImg1.shape[0], gImg1.shape[1]))
```

```python
for h in range(0, gImg1.shape[0]):
    for w in range(0, gImg1.shape[1]):
        val = gImg1[h, w] % 64
        val = val * 4
        gOutimg[h, w] = val

plt.imshow(gOutimg, cmap='gray')
```

```python
img2 = cv2.imread('../images/lena256.jpg')
RGB_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

gImg2 = cv2.cvtColor(RGB_img2, cv2.COLOR_RGB2GRAY)
print(gImg2.shape)
plt.imshow(gImg2, cmap='gray')
```

```python
gOutimg2 = np.zeros((gImg2.shape[0], gImg2.shape[1]))
```

```python
for h in range(0, gImg2.shape[0]):
    for w in range(0, gImg2.shape[1]):
        val = gImg2[h, w] % 64
        val = val * 4
        gOutimg2[h, w] = val

plt.imshow(gOutimg2, cmap='gray')
```

```python
equalVal = 200
width = 5
min = equalVal - width
max = equalVal + width
```

```python
for h in range(0, gImg2.shape[0]):
    for w in range(0, gImg2.shape[1]):
        if (gOutimg2[h, w] > min and gOutimg2[h, w] < max):
            gOutimg2[h, w] = 0
        else:
            gOutimg2[h, w] = gImg2[h, w]

plt.imshow(gOutimg2, cmap='gray')
```

```python
for h in range(0, gImg2.shape[0]):
    for w in range(0, gImg2.shape[1]):
        val = 255*(gImg2[h, w] / 128 - 1)**2
        gOutimg2[h, w] = val

plt.imshow(gOutimg2, cmap='gray')
```

```python
for h in range(0, gImg2.shape[0]):
    for w in range(0, gImg2.shape[1]):
        imVal = (int)(gImg2[h, w] / 128)
        if(imVal == 0):
            gOutimg2[h, w] = imVal
        else:
            gOutimg2[h, w] = 255


plt.imshow(gOutimg2, cmap='gray')
```

```python
planeNumber = 2
planeValue = 1

for i in range(planeNumber-1):
    planeValue = planeValue * 2
```

```python
for h in range(0, gImg2.shape[0]):
    for w in range(0, gImg2.shape[1]):
        imVal = (int)(gImg2[h, w] / planeValue)
        if(imVal % 2 == 0):
            gOutimg2[h, w] = 0
        else:
            gOutimg2[h, w] = 255


plt.imshow(gOutimg2, cmap='gray')
```
